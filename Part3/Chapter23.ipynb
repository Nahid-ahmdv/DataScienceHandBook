```python
import yfinance as yf

# Download data
sp500 = yf.download('^GSPC', start='2018-01-01', end='2022-12-31')

# Drop the 'Ticker' level from the column MultiIndex
sp500.columns = sp500.columns.droplevel('Ticker')

# Reorder columns
sp500_main = sp500[['High', 'Low', 'Open', 'Close', 'Volume']]
sp500_main.head()

```

```python
sp500 = sp500['Close']
sp500
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
sp500.plot();
```

```python
sp500.plot(alpha=0.5, style='-')
sp500.resample('BYE').mean().plot(style=':')
sp500.asfreq('BYE').plot(style='--');
plt.legend(['input', 'resample', 'asfreq'],
loc='upper left');
```

```python
# From daily to monthly, taking the mean of each month
sp500.resample('ME').mean()
```

```python
sp500.iloc[:4]
```

```python
fig, ax = plt.subplots(2, sharex=True)
data = sp500.iloc[:20]
data.asfreq('D').plot(ax=ax[0], marker='o')
data.asfreq('D', method='bfill').plot(ax=ax[1], style='-o')
data.asfreq('D', method='ffill').plot(ax=ax[1], style='--o')
ax[1].legend(["back-fill", "forward-fill"]);
```

```python
sp500 = sp500.asfreq('D', method='pad')
ROI = 100 * (sp500.shift(-365) - sp500) / sp500
ROI.plot()
plt.ylabel('% Return on Investment after 1 year');
```

```python
sp500.plot()
```

```python
rolling = sp500.rolling(365, center=True)
data = pd.DataFrame({'input': sp500, 'one-year rolling_mean': rolling.mean(), 'one-year rolling_std': rolling.std()})
ax = data.plot(style=['-', '--', ':'])
ax.lines[0].set_alpha(0.3)
```

```python
rolling = sp500.rolling(365, center=True)
data = pd.DataFrame({'input': sp500, 'one-year rolling_mean': rolling.mean(), 'one-year rolling_median': rolling.median()})
ax = data.plot(style=['-', '--', ':'])
ax.lines[0].set_alpha(0.3)
```

## <b>Example: Visualizing Seattle Bicycle Counts</b>

```python
url = 'https://raw.githubusercontent.com/jakevdp/bicycle-data/main/FremontBridge.csv'
data = pd.read_csv(url, index_col='Date')
data.head()
```

```python
data.columns = ['Total', 'East', 'West']
data.index = pd.to_datetime(data.index)
data.head()
```

```python
data.index.time
```

```python
data.plot()
plt.ylabel('Hourly bicycle count')
```

```python
# import pandas as pd
# import matplotlib.pyplot as plt
import matplotlib.dates as mdates #This imports the dates module from matplotlib, which provides utilities for working with dates and times, such as formatting and tick positioning on date-based axes.

# Assuming 'data' is your DataFrame with Date as index
data.index = pd.to_datetime(data.index)
'''
This ensures that the index of the 'data' DataFrame is treated as 'datetime' objects (if it's not already). 
The pd.to_datetime() function converts the index (which should represent date or time) into a proper 'datetime' format for plotting.
'''

ax = data.plot()
'''
This generates a line plot of the data. The 'data.plot()' function automatically uses the DataFrame's index (which, after the previous step, is in a datetime format) on the x-axis, and the values are plotted on the y-axis.
'''
plt.ylabel('Hourly bicycle count')

# Format x-axis as 'YYYY-MM'
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
'''
The .set_major_formatter() function is used to define how the major ticks on an axis should be formatted in a plot. In the context of plotting with matplotlib, axes have ticks that correspond to the positions on the axis, and these ticks can be labeled in various formats (such as numbers, dates, or custom labels).
This formats the x-axis to display the date in the YYYY-MM format (e.g., "2025-04"). The mdates.DateFormatter('%Y-%m') specifies the desired format for the date labels on the x-axis.
'''
plt.xticks(rotation=45)
'''
This rotates the x-axis tick labels by 45 degrees to ensure they are readable, especially when the dates are close together, like when plotting data by month.
'''
plt.show()
```

```python
weekly = data.resample('W').sum()
weekly.plot(style=[':', '--','-'])
plt.ylabel('Weekly bicycle count');
```

```python
daily = data.resample('D').sum()
daily.rolling(30, center=True).sum().plot(style=['-', ':', '--'])
plt.ylabel('mean hourly count');
```

```python
daily.head()
```

```python
daily.rolling(50, center=True,
win_type='gaussian').sum(std=10).plot(style=['-', ':', '--']);
```

```python
by_time = data.groupby(data.index.time).mean()
hourly_ticks = 4 * 60 * 60 * np.arange(6)
by_time.plot(xticks=hourly_ticks, style=['-', ':', '--']);
```

```python
by_time.head()
```

```python
by_weekday = data.groupby(data.index.dayofweek).mean()
by_weekday.index = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
by_weekday.plot(style=['-', ':', '--']);
```

```python
data['Total'] = data['East'] + data['West']
data['Total']= data.eval('East + West')
data.dropna().describe()
```

```python
(data['Total'] == data['East'] + data['West']).all()
'''
 What it tells you:
This checks row by row: is Total == East + West?

Then .all() returns:

True   every row matches

False   at least one row doesn't match
'''
```

```python
data[data['Total'] != data['East'] + data['West']]
# .shape    (23, 3)
```

```python
weekend = np.where(data.index.weekday < 5, 'Weekday', 'Weekend')
by_time = data.groupby([weekend, data.index.time]).mean()
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
by_time.loc['Weekday'].plot(ax=ax[0], title='Weekdays',
xticks=hourly_ticks, style=['-', ':', '--'])
by_time.loc['Weekend'].plot(ax=ax[1], title='Weekends',
xticks=hourly_ticks, style=['-', ':', '--']);
```
